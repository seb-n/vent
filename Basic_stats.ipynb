```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load user data
personals_df = pd.read_csv("node_attributes.csv")
#convert dtypes
personals_df.loc[:, personals_df.columns!="gender"] = personals_df.drop("gender", axis=1).astype(int)
```

### Age distribution

```python
sns.histplot(personals_df[(personals_df["age"]>0) & (personals_df["age"]<100)]["age"], stat="probability", bins=100)
plt.xlabel("age")
plt.tight_layout()
plt.savefig("age_dist")
```

### Gender distribution

```python
personals_df["gender"].value_counts()
```

### Posts created

```python
post_vc = personals_df.post_count.value_counts()
plt.scatter(x=post_vc.index, y=post_vc.values/post_vc.sum(), alpha=0.2)
plt.xscale("log")
plt.yscale("log")
plt.ylabel("frequency")
plt.xlabel("post count")
plt.savefig("post_count")
```

### Time on platform

```python
sns.histplot(personals_df["time_on_platform"], stat="probability", bins=100)
plt.tight_layout()
plt.savefig("time_on_platform")
```
